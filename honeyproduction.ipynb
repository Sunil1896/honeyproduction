```python
years = [2019, 2020, 2021, 2022]
data = [25, 30, 20, 25]  

plt.figure(figsize=(8, 8))
plt.pie(data, labels=years, autopct='%1.1f%%', startangle=140)
plt.title('Percentage Distribution of Data by Year')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='numcol', y='prodvalue', data=df)
plt.title('Relationship between numcol and prodvalue')
plt.xlabel('Number of Colonies')
plt.ylabel('Production Value')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='year', y='prodvalue', data=df)
plt.title('Relationship between Year and Production Value')
plt.xlabel('Year')
plt.ylabel('Production Value')
plt.show()
```

```python
selected_columns = ['numcol', 'yieldpercol', 'totalprod', 'prodvalue', 'year']
df1 = df[selected_columns]

sns.pairplot(df1, hue='year', diag_kind='hist')
plt.suptitle('Pairplot of Variables by Year', y=1.02)
plt.show()
```

```python
selected_columns = ['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue']
df2 = df[selected_columns]

correlation_matrix = df2.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 10})
plt.title('Correlation Matrix')
plt.show()
```
